#  Basic Line Plot,Generating x and y values

#Plot a simple line graph of ùë¶ = ùë•^2

for ùë• values ranging from -10 to 10.

Label the x-axis as "X Values" and the y-axis as "Squared Values."

 Add a title to the graph.

```python
import numpy as np
import matplotlib.pyplot as plt # Taking plot library
import seaborn as sns
import random
x = np.arange(-10, 11)
print(x)
y = x**2
print(y)
```

# Plotting the line graph

```python
plt.figure()
plt.plot(x, y)
plt.xlabel("X Values")
plt.ylabel("Squared Values")
plt.title("Line Plot of y = x^2")
plt.show()
```

# Scatter Plot

 Generate two lists of 20 random integers each (for x and y coordinates).

 Create a scatter plot with these values, and customize it by adding color, size, and labels to the points.

```python
x_random = random.sample(range(1, 100), 20)
y_random = random.sample(range(1, 100), 20)
sizes = np.random.randint(50, 200, size=20)
colors = np.random.rand(20)
plt.figure()
plt.scatter(x_random, y_random, s=sizes, c=colors, cmap='viridis', alpha=0.7)
plt.colorbar(label="Color Scale")
plt.xlabel("X Random Values")
plt.ylabel("Y Random Values")
plt.title("Scatter Plot of Random Values")
plt.show()
```

#Histogram

Simulate 100 random values from a normal distribution using np.random.randn() and plot a histogram.

Customize the number of bins and add a grid to the plot.

```python
random_values = np.random.randn(100)
plt.figure()
plt.hist(random_values, bins=10, color='purple', alpha=0.7, edgecolor='black')
plt.title("Histogram of Normally Distributed Values")
plt.grid(True)
plt.show()
```

#Pie Chart

Visualize the percentage distribution of five categories ['A', 'B', 'C', 'D', 'E'] with values [20, 30, 25, 15, 10].

 Display the percentage values on the chart and use a color palette for differentiation.

```python
categories = ['A', 'B', 'C', 'D', 'E']
values = [20, 30, 25, 15, 10]
colors = ['black', 'red', 'lightgreen', 'pink', 'orange']
plt.figure()
plt.pie(values, labels=categories, autopct="%1.1f%%", colors=colors)
plt.title("Percentage Distribution of Categories")
plt.show()
```

#Comparison of Matplotlib and Seaborn
Generate a pair of plots (bar charts or line plots) using the same data in both Matplotlib and Seaborn.

Highlight the differences in the code, syntax, and aesthetics of the two libraries

#Matplotlib Bar Chart

```python
import matplotlib.pyplot as plt
import numpy as np
students = ['Amaira', 'Bella', 'Casie', 'Derek', 'Emerlad']
subjects = ['Math', 'Science', 'English']
scores = np.random.randint(50, 100, (5, 3))
fig, ax = plt.subplots(figsize=(8, 6))
width = 0.2
x = np.arange(len(students))

for i, subject in enumerate(subjects):
    ax.bar(x + i * width, scores[:, i], width, label=subject)
ax.set_xlabel('Students')
ax.set_ylabel('Scores')
ax.set_title('Student Scores by Subject (Matplotlib)')
ax.set_xticks(x + width)
ax.set_xticklabels(students)
ax.legend()

# Display the plot
plt.show()
```
